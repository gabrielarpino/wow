```python
import sympy
from sympy import Piecewise, log, ITE, piecewise_fold, exp
from sympy.abc import x, y, z, w, k, j
from sympy import *
from sympy.plotting import plot
```

```python
psi = Piecewise((1, (0 <= x) & (x < 0.5)), (-1, (0.5 <= x) & (x < 1)), (0, True))
f = exp(-abs(x))
psi
_psi = 2**(j/2) * psi.subs(x, 2**j * y - k)
# yo = lambda y: psi.subs(x, y)
```

```python
w_coeffs = []
for i in range(10):
    for q in range(10):
        _psi = 2**(j/2) * psi.subs([(x, 2**j * x - k)])
        p = _psi.subs([(j, i), (k, q)]) * f
        w_coeffs.append(integrate(p, (x, -oo, oo)))

```
